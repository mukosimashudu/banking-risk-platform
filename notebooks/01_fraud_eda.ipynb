#Import


```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

#Data Ingestion — Load Raw IEEE Fraud Dataset

```python
fraud_file = Path("/content/train_transaction.csv")
df = pd.read_csv(fraud_file)
df.head()
```

#Inspect Dataset Structure

```python
# Check number of rows and columns
df.shape
# Display column info and data types
df.info()
```

##Check Missing Values

```python
# Calculate missing values per column
df.isnull().sum()
```

##Analyze Fraud Distribution

```python
# Count fraud vs non-fraud cases
df['isFraud'].value_counts()
```

##Fraud Distribution (Proportion Analysis)

```python
df["isFraud"].value_counts(normalize=True)
```

##Explore Transaction Amount

```python
# Summary statistics for transaction amount
df['TransactionAmt'].describe()
```

#cell 6

```python
df["TransactionAmt"].hist(bins=50)
plt.title("Transaction Amount Distribution")
plt.xlabel("TransactionAmt")
plt.ylabel("Frequency")
plt.show()
```
